**Reading diffeent data sources with the help of pandas**

->When a csv file is read it is converted into data frames.This is done using **read_csv*

->**StringIO** is a class in the io module that lets you treat a string like a file object.For eg: If we have a csv data as a string it will convert string into a file like object then we read the data using** pd.read_csv()**

->**in-memory file format** it means you’re working with file-like data stored in memory (RAM) instead of saving or reading it from a physical file on disk.

In above data Column d is of type float64 because it contains at least one missing value (NaN). Here's why that matters:

In pandas, missing values (NaN) are represented as float by default.

So, even if the other values in column d are integers, the presence of a NaN forces pandas to convert the entire column to float64 to accommodate it.

```python
data = '{"employee_name":"James","email":"james@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Sr.Developer"}]}'
```

```python
type(data)
```

```python
pd.read_json(data)
```

```python
pd.read_json(data,orient='records')  #orient='records' tells pandas how the JSON is structured — in this case, it's a list of records (dictionaries). orient='column' also give the same output
```

```python
pd.read_json(data,orient='index')
```

```python
pd.read_json(data,orient='split')
#we cannot use spit with this format of json
#Format for orient='split'
#It should be a dictionary with 3 specific keys:

#"columns" → a list of column names

#"index" → a list of row indices

#"data" → a 2D list of values (rows)
#eg:{
  "columns": ["name", "age"],
  "index": [0, 1, 2],
  "data": [
    ["Alice", 25],
    ["Bob", 30],
    ["Charlie", 28]
  ]
}

```

```python
data = [
    {
        "id": 1,
        "name": "Cole Volk",
        "fitness": {"height": 130, "weight": 60},
    },
    {"name": "Mark Reg", "fitness": {"height": 130, "weight": 60}},
    {
        "id": 2,
        "name": "Faye Raker",
        "fitness": {"height": 130, "weight": 60},
    },
]
```

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data,max_level=0) # max_level=0 means we have to go only upto one level of nested dictionary
```

```python
data = [
    {
        "state": "Florida",
        "shortname": "FL",
        "info": {"governor": "Rick Scott"},
        "counties": [
            {"name": "Dade", "population": 12345},
            {"name": "Broward", "population": 40000},
            {"name": "Palm Beach", "population": 60000},
        ],
    },
    {
        "state": "Ohio",
        "shortname": "OH",
        "info": {"governor": "John Kasich"},
        "counties": [
            {"name": "Summit", "population": 1234},
            {"name": "Cuyahoga", "population": 1337},
        ],
    },
]
```

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data, "counties",["state", "shortname",["info",'governor']])
```
