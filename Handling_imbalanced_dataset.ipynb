**Imbalanced dataset:**

A dataset with an unequal class distribution.

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas dataframe
credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0-->legit transactions

1-->fraudulent transactions

```python
# seperating the legit and fraudulent transactions
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Under-sampling**

build a sample dataset containing similar distribution of legit and fraudulent transactions.

```python
legit_sample=legit.sample(n=211)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

axis=0 means first we will have legit_sample rows and then we will have fraud rows.
