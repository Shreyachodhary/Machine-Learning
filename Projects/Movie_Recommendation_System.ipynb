Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib  #it is used to get the closest match of the some value
from sklearn.feature_extraction.text import TfidfVectorizer # convert textual data into numerical values
from sklearn.metrics.pairwise import cosine_similarity
```

cosine_similarity is used to measure how similar two vectors are regardless of their magnitude — it looks at the angle between the vectors rather than their distance.



Data Collection and Pre-Processing

```python
#loading the data from the csv file to a pandas frame
movies_data=pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# selecting the relevant features for recommendation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string

for features in selected_features:
  movies_data[features]=movies_data[features].fillna('')
```

```python
#combining all the 5 selected features
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

Cosine Similarity

```python
#getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting the movie name from the user

```python
#getting the movie name from the user

movie_name=input('Enter your favourite movie name:')
```

```python
#converting a list with all the movie names given in the dataset
list_of_all_titles=movies_data['title'].tolist()  #tolist will take all the values and create a list
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title

index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies

similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

similarity_score: This is expected to be a list of items (most likely tuples or lists), where each item contains at least two elements.

key=lambda x: x[1]: This tells the sorted() function to sort the items by the second element of each item (i.e., x[1]).

reverse=True: Sorts the list in descending order.

```python
# print the name of similar movies based on the index
print('Movies suggested for you : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  print(title_from_index)
  if(i<20):
    print(i,'.',title_from_index)
    i+=1
```

**Movie Recommendation System**

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
