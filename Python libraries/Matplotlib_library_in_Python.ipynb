**Matplotlib**:useful for making Plots



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100) # wil generate evenly spaced 100 values between 0 and 10
y=np.sin(x)
z=np.cos(x)
```

```python
print(y)
```

**Plotting the data**

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title ,x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')  # give a title to our graph
plt.show()
```

```python
# parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')  # 'r+': Format string that specifies the style of the plot.
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

**How to plot different lines**

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

# **Bar Plot**: helps in Visualize class distributions (e.g., how many samples per category in classification).

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)    # creates a vertical bar chart
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

🔹 fig = plt.figure()
Creates a new figure — like a blank canvas where your plot(s) will go.

🔹 ax = fig.add_axes([0, 0, 1, 1])
Adds an Axes (the actual plot area) to the figure.

The list [0, 0, 1, 1] specifies:

left = 0 (start from left edge of the figure),

bottom = 0 (start from bottom edge),

width = 1 (take 100% of width),

height = 1 (take 100% of height).

So this fills the entire figure with one plot.

# **Pie Chart**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

->ax.pie(people, labels=languages, autopct='%1.1f%%'):
This creates a pie chart using the values from people.

Breakdown:
people: The data (sizes of each slice).

labels=languages: Sets labels for each slice.

autopct='%1.1f%%': Displays the percentage on each slice, formatted to 1 decimal place (like 35.7%).

# **Scatter Plot** : in a scatter plot data point are scattered so there wonts be any line joining the funtions and it is very useful in clustering application

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])  # rectangular axis
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

# **3D Scatter Plot**

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')  # projection='3d' part is what makes the plot three-dimensional.
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

->z = 20 * np.random.random(100)::
Generates 100 random values between 0 and 20 for the Z-axis.

np.random.random(100) gives 100 values in the range [0, 1), and multiplying by 20 scales it to [0, 20).



**->ax.scatter(x, y, z, c=z, cmap='Blues')::**
This line creates a 3D scatter plot!

Let’s decode the arguments:

Argument
x, y, z	3D coordinates for each point

c=z	Sets the color of each point based on its z-value

cmap='Blues'	Applies the Blues colormap — points with lower z will be light blue, higher z will be darker
