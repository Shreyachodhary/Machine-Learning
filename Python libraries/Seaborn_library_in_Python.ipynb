```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Note** : Seaborn has some built in datasets

```python
# total bill vs tip dataset
tips=sns.load_dataset('tips')  # this will be imported in the form of pandas dataframe
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

>**replot()** allows to split the vizualization into multiple subplots(called facets) using row or col.


data=tips	Uses the tips dataset (e.g., from sns.load_dataset("tips"))

x='total_bill'	X-axis is total bill amount

y='tip'	Y-axis is tip amount

col='time'	Creates separate plots (columns) for Lunch and Dinner times

hue='smoker'	Points are colored based on whether the person is a smoker or not

style='smoker'	Also gives different markers (e.g. circle, triangle) for smokers vs non-smokers

size='size'	Point size is based on party size (how many people were in the group)

**Correlation with Heatmap**

A correlation heatmap uses colored cells,typically in a monochromatic scale,to show a 2D correlation matrix (table) between two discrete dimensions or event types.It is very important in Feature selection.

whenever we are applying correlation your value will be ranging between -1 to 1.

```python
df.select_dtypes(include='number').corr()  # we have to use this command because df contains both categoral columns and numeric data
```

```python
sns.heatmap(df.select_dtypes(include='number').corr())
```

**JoinPlot**

A join plot allows to study the relationship between 2 numeric variables.The central chart display their correlation . It is usually a scatterplot , a hexbin plot ,a 2D histogram or a 2D density plot.

**Univariate Analysis**

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

**kind='hex':**

This tells Seaborn to use a hexbin plot instead of a scatterplot.

A hexbin plot divides the plot area into hexagonal bins and counts how many data points fall into each bin.

It's especially useful when you have a large number of data points and want to see density or clustering.

**Top and right histograms**: these show the distribution of the individual variables ('tip' and 'total_bill') separately.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

kind='reg':

This tells Seaborn to plot a scatterplot with a regression line.
>Plot the individual data points (scatterplot).

>Draw a linear regression line that best fits the data.The regression line helps you see whether there's a positive or negative trend.

**Pair Plot**

A pair plot creates a matrix of plots to show pairwise relationships between multiple variables in a dataset.

Instead of comparing just two variables like jointplot, pairplot does it for all combinations — perfect for small to medium datasets.

Plot scatterplots for every pair of numerical columns.

```python
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
sns.pairplot(df,hue='sex')
```
